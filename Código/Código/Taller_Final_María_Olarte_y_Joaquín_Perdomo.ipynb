```python
# modelo logistico
from sklearn.linear_model import LogisticRegression
y['Total_predicted'] = y['G1'] + y['G2'] + y['G3']
y['Aprobado'] = (y['Total_predicted'] >= 40).astype(int)
X_log = x
y_log = y['Aprobado']

X_train_log, X_test_log, y_train_log, y_test_log = train_test_split(X_log, y_log, test_size=0.2, random_state=42)

log_model = LogisticRegression()
log_model.fit(X_train_log, y_train_log)
y['Aprobado_prediccion'] = log_model.predict(X_log)

y['Aprobado'] = y['Aprobado'].map({1: 'Aprobado', 0: 'No aprobado'})
y['Aprobado_prediccion'] = y['Aprobado_prediccion'].map({1: 'Aprobado', 0: 'No aprobado'})
print(y[['G1', 'G2', 'G3', 'Total_predicted', 'Aprobado', 'Aprobado_prediccion']])

print(y[y['Aprobado_prediccion'] == 'Aprobado'])
```

```python
from sklearn.metrics import roc_curve, roc_auc_score, confusion_matrix, classification_report
cm = confusion_matrix(y['Aprobado'], y['Aprobado_prediccion'])
print('\nMatriz de Confusión:')
print(cm)
```
